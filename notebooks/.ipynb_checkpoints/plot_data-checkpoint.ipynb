```python
import ee
import pandas as pd
import numpy as np
import geopandas as gp
import matplotlib.pyplot as plt
%matplotlib inline
ee.Initialize()
```

## Plot crop ET time series for each crop by month

```python
df = pd.read_csv('../data/et_crops_2001_2017.csv', index_col=0, parse_dates=True)
f,ax=plt.subplots(1,figsize=(10,4))
df.plot(ax=ax)
ax.set_ylabel('ET [mm/month]')
plt.legend(loc='upper left')
```

## Plot crop ET time series for each year

```python
df = pd.read_csv('../data/et_crops_2001_2017.csv', index_col=0, parse_dates=True)
f,ax=plt.subplots(1,figsize=(10,4))
df.resample('A').sum().plot(ax=ax, marker='o')
ax.set_ylabel('Annual ET [mm]')
plt.legend(loc='upper left')
plt.title('Annual total ET')
```
